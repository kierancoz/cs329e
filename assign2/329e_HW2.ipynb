## Create a vector of 1000 random numbers from a Normal distribution with mean 5 and standard deviation 10 and plot the discrete probability density function using a histogram

```python
x = np.random.normal(5,10,1000)
plt.hist(x, density = True, bins = 20)
plt.title("Discrete Probabiltiy Function")
plt.xlabel("Number")
plt.ylabel("Probability")
plt.show()
```

Now, using matrix multiplication rotate the vectors 45degress and draw the result

```python
matrix1 = np.array([[0,1], [0,1]])
matrix2 = np.array([[1,0], [1,1]])
plt.plot(matrix1[0], matrix1[1])
plt.plot(matrix2[0], matrix2[1])
plt.show()
```

```python
theta = math.radians(45)
rotation_matrix = np.array([[math.cos(theta), math.sin(theta)], [-math.sin(theta), math.cos(theta)]])

# Counter-clockwise rotation is as follows
new_matrix1 = np.transpose(np.transpose(matrix1).dot(rotation_matrix))
new_matrix2 = np.transpose(np.transpose(matrix2).dot(rotation_matrix))

# Clockwise rotation would be as follows
#new_matrix1 = rotation_matrix.dot(matrix1)
#new_matrix2 = rotation_matrix.dot(matrix2)

plt.plot(new_matrix1[0], new_matrix1[1])
plt.plot(new_matrix2[0], new_matrix2[1])
plt.axis('square')
plt.show()
```
